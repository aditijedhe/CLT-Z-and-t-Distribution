# Central Limit Theorem

The Central Limit Theorem states that if you select sufficiently **large random samples** from a population with mean μ and standard deviation σ, then the sample means will be approximately follow Normal Distribution with mean μ and standard deviation σ2/n, irrespectively of the distribution of the population.

## Implementing Central Limit Theorem in Python
We will be demostrating the above theorem using python

```python
# importing necessary librarires
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import statistics as stats
import random
from scipy.stats import norm
```

```python
# building uniform distribution as population

a,b= -19.68, 28   # giving the parameters for the distribution
s = np.random.uniform(a, b, 100000)
```

```python
# a list for sample mean
meansample = [] 

# number of sample
numofsample = [100,250,500,1000,2500,5000] 

# sample size
samplesize =50  

# iterating for each number of sampling (100 to 5000)
for i in numofsample: 
    # collect mean of each sample
    
    eachmeansample = [] 
    # iterating for each sample value 
    for j in range(0,i):
        
        # sampling 500 sample from population
        rc = random.choices(s, k=samplesize) 
        
        # collect mean of each sample
        eachmeansample.append(sum(rc)/len(rc)) 
        
    # appending the mean of each sampling to the list
    meansample.append(eachmeansample) 
```

```python
# assigning the position for the plot 
cols = 2
rows = 3

#plotting the distribution of the mean of the six sample
fig, ax = plt.subplots(rows, cols, figsize=(20,15)) 
n = 0

#iterating over the positions of the subplots
for i in range(0, rows):
    for j in range(0, cols):
        # plotitng histogram for mean of each sample
        ax[i, j].hist(meansample[n], 200, density=True, color="brown") 
        
        # giving the title to each figure
        ax[i, j].set_title(label="number of sampling :" + str(numofsample[n])) 
        n += 1 
plt.show()
```

From the digram we can interpret that when n is sufficiently large i.e., greater than 250 we see that the sample mean tends tends to follow Normal Distribution, this is exactly what our Central Limit Theorem states.

# Normal Distribution Curve

Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.

```python
# Plot between -1995 and 200 with .025 steps
x = np.arange(-1995, 200, 0.025)
  
# Calculating mean and standard deviation
mean = stats.mean(x)
sd = stats.stdev(x)

# plotting the distribution curve
plt.plot(x, norm.pdf(x, mean, sd), color="purple") 

# drwaing the line passing through the mean
plt.axvline(mean ,color="black", linestyle="dashed") 

#Fill under the curve
plt.fill_between(
        x= x, 
        y1= norm.pdf(x, mean, sd),
        color= "purple",
        alpha= 0.2)
        
plt.show()
```

The dotted line represents the mean of the data, where most of the datapoints are densed, the density of the data decreases as it moves away from the mean. 

# Standard Normal Distribution

The standard normal distribution is one of the forms of the normal distribution. It occurs when a normal random variable has a mean equal to zero and a standard deviation equal to one. In other words, a normal distribution with a mean 0 and standard deviation of 1 is called the standard normal distribution. Also, the standard normal distribution is centred at zero, and the standard deviation gives the degree to which a given measurement deviates from the mean.<br>


The random variable of a standard normal distribution is known as the standard score or a **z-score**. It is possible to transform every normal random variable X into a z score using the following formula:<br>

**z = (X – μ) / σ**<br>

where X is a normal random variable, μ is the mean of X, and σ is the standard deviation of X.<br>

# t Distribution Curve

t-distribution looks similar since it is also centered at zero and has a bell-shape. Still, it is shorter, flatter and its standard deviation is proportionally larger compared with the Z-distribution. The t distribution is sensitive of sample size and is used for **small samples** when the population **standard deviation is unknown** .

```python
from scipy.stats import t
t_dist = t.rvs(df=2, size=40) # generating an sample with 12 degrees of freedom and sample size=1000
sns.kdeplot(t_dist, shade=True, color="g") # plogts the distribution curve
plt.axvline(x=t_dist.mean(), color="black",linestyle="dashed") # to draw the mean line
 
plt.show()
```

The dotted line represents the mean of the data, where most of the datapoints are densed.
